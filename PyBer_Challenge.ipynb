# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
ridesTotal = pyber_data_df.groupby(['type']).count()['ride_id']
ridesTotal
```

```python
# 2. Get the total drivers for each city type
driversTotal = pyber_data_df.groupby(['type']).sum()['driver_count']
driversTotal
```

```python
#  3. Get the total amount of fares for each city type
faresTotal = pyber_data_df.groupby(['type']).sum()['fare']
faresTotal
```

```python
#  4. Get the average fare per ride for each city type. 
faresAvg_perRide = faresTotal/ridesTotal
faresAvg_perRide
```

```python
# 5. Get the average fare per driver for each city type. 
faresAvg_perDriver = faresTotal/driversTotal
faresAvg_perDriver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides":ridesTotal,
    "Total Drivers":driversTotal,
    "Total Fares":faresTotal,
    "Average Fare per Ride":faresAvg_perRide,
    "Average Fare per Driver":faresAvg_perDriver
                                })
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)


#source: https://pandas.pydata.org/docs/reference/api/pandas.io.formats.style.Styler.format.html
pyber_summary_df = pyber_summary_df.style.format(precision=2, thousands=',')


pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly [sic] of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_summary_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
faresSummed_df = pyber_data_df.groupby(['type','date']).sum()['fare']
faresSummed_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. 
#This is needed to use the 'pivot()' function.

faresSummed_df = faresSummed_df.reset_index()
faresSummed_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
faresSummed_df_Pivot = faresSummed_df.pivot(index='date', columns='type',values='fare')
faresSummed_df_Pivot
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame \
#using loc on the given dates, '2019-01-01':'2019-04-28'.

faresSummed_df_Pivot = faresSummed_df_Pivot.loc['2019-01-01':'2019-04-28']
faresSummed_df_Pivot
```

```python
# 5. Set the "date" index to datetime datatype. 
#This is necessary to use the resample() method in Step 8.
faresSummed_df_Pivot.index = pd.to_datetime(faresSummed_df_Pivot.index)
faresSummed_df_Pivot
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
faresSummed_df_Pivot.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function 
# by week 'W' and get the sum of the fares for each week.
faresSummed_df_Pivot_perWeek = faresSummed_df_Pivot.resample("W").sum()
faresSummed_df_Pivot_perWeek
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame 
# using the df.plot() function. 
fig, ax = plt.subplots(figsize=(15,5))
faresSummed_df_Pivot_perWeek.plot(ax=ax)
ax.set_xlabel("")
ax.set_ylabel("Fare ($USD)")
ax.set_title("Total Fare by City Type")

#source:https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html
ax.legend(loc='center')

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

```
